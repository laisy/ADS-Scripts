```python
tempos = []
for k in lines:
    date = k[0] + ' ' + k[1]
    date = datetime.strptime(date, '%Y-%m-%d %H:%M:%S')
    hora = date.time()

    tempos.append(date)
 
tempo_processo = tempos[-1] - tempos[0]
print('TEMPO PROCESSO PYTHON SEM CARGA:', tempo_processo)
```

```python
cpus = []
for k in lines:
    cpus.append(float(k[4]))
 
cpus_utilizada =  "{:.2f}".format(sum(cpus) / len(cpus))
print('%CPU PROCESSO PYTHON COM CARGA:', cpus_utilizada, "%")
```

```python
memorias = []
for k in lines:
    memorias.append(float(k[5]))
 
memorias_utilizada =  "{:.2f}".format(sum(memorias) / len(memorias))
print('%MEM PROCESSO PYTHON COM CARGA:', memorias_utilizada, "%")
```

```python
memorias = []
for k in lines:

    # MEMTOTAL bytes
    mem_value = (float(k[5])*int(k[3]))/100
    memorias.append(mem_value)
 
memorias_utilizada = sum(memorias) / len(memorias)
memorias_utilizada = "{:.2f}".format(memorias_utilizada/1024)

print('MEM PROCESSO PYTHON COM CARGA:', memorias_utilizada, 'KB')
```
