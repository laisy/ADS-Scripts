```python
lines = []
for line in open('/home/laisy/Documentos/ADS/ADS-Scripts/Scripts-Python/monitoramento_py.txt', 'r'):
    lines.append(line.split())

```

```python
x = []
y = []
for k in lines:
    date = k[1] + ' ' + k[2]
    date = datetime.strptime(date, '%Y/%m/%d %H:%M:%S')
    hora = date.time()
    # MemUsed / MemTotal
    MemUsed = round((float(k[10]) / float(k[8]))*100, 1)
    if MemUsed not in y:
        y.append(MemUsed)
```

```python
cont = 0
for k in y:
    cont += 1
    x.append(cont)
```

```python
plt.title("Qtd Memria X Segundos")
plt.xlabel('Segundos')
plt.ylabel('Memria Usada %')
plt.plot(x, y, marker = 'o', c = 'g')
# plt.ylim(50, 75)
plt.show()
```
